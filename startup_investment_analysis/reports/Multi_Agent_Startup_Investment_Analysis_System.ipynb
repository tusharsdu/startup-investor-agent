## üèóÔ∏è Agentic Layer - System Architecture

### What is the Agentic Layer?
The agentic layer is the core orchestration system that manages communication between all AI agents using the Model Context Protocol (MCP). Each agent is specialized for a specific analysis domain while maintaining standardized interfaces for seamless integration.

### Agent Roles & MCP Interactions:

**üîÄ MCP Router Agent** (Central Orchestrator)
- **Purpose**: Routes requests to appropriate agents and manages workflows
- **MCP Role**: Primary message router and coordinator
- **Interactions**: Communicates with all agents, manages parallel processing
- **Workflow**: Receives pitch deck ‚Üí distributes to analysis agents ‚Üí collects results

**üìù Summarization Agent** 
- **Purpose**: Creates executive summaries of pitch decks
- **MCP Role**: Content processor and synthesizer
- **Interactions**: Receives raw pitch text from MCP Router
- **Output**: Structured JSON summary stored in FAISS

**üí∞ Financial Analysis Agent**
- **Purpose**: Analyzes financial metrics, projections, and investment potential
- **MCP Role**: Financial data processor with calculation engine
- **Interactions**: Processes financial sections via MCP Router
- **Storage**: Financial analysis vectors stored in FAISS for Q&A retrieval

**‚ö†Ô∏è Risk & Compliance Agent**
- **Purpose**: Assesses regulatory, legal, and operational risks
- **MCP Role**: Risk evaluation specialist
- **Interactions**: Receives compliance-related queries from MCP Router
- **Output**: Risk scores and mitigation strategies

**üîß Technical Analysis Agent**
- **Purpose**: Evaluates technology stack, scalability, and innovation
- **MCP Role**: Technical feasibility assessor
- **Interactions**: Routes technical questions through MCP
- **Analysis**: Architecture, scalability, and technical debt assessment

**üìä Market Analysis Agent**
- **Purpose**: Comprehensive market research and competitive analysis
- **MCP Role**: Market intelligence processor with web scraping
- **Data Sources**: Bing API integration via MCP Router
- **Storage**: Market vectors in FAISS for intelligent retrieval

**‚ùì VC Question-Answer Agent**
- **Purpose**: Autonomous question generation and intelligent routing
- **MCP Role**: Question orchestrator and response coordinator
- **Workflow**: Generates 10 VC questions ‚Üí Routes via MCP ‚Üí Retrieves from FAISS
- **Intelligence**: Uses retriever pattern for context-aware responses

**‚öñÔ∏è Investment Verdict Agent**
- **Purpose**: Final investment decision synthesis
- **MCP Role**: Decision aggregator and verdict generator
- **Input**: All agent analyses + Q&A results
- **Output**: Final investment recommendation with reasoning

### Base Agent Class - Foundation for All Agents
**From: `core/base_agent.py` - Your exact implementation**

````python
# BaseAgent: EXACT code from core/base_agent.py
from abc import ABC, abstractmethod

class BaseAgent(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def process(self, input_data):
        pass

print("‚úÖ BaseAgent loaded exactly from your workspace: core/base_agent.py")
print("üìã This is the foundation class that all agents inherit from")
````

````python
# SummarizationAgent: EXACT code from agents/summarization_agent.py
from openai import OpenAI

class SummarizationAgent(BaseAgent):
    def __init__(self, perplexity_api_key: str):
        super().__init__("SummarizationAgent")
        self.client = OpenAI(
            api_key=perplexity_api_key,
            base_url="https://api.perplexity.ai"
        )

    def process(self, pitch_text: str) -> dict:
        prompt = f"""
You are a senior venture capital analyst with 10+ years of experience reviewing startup pitch decks. Create a comprehensive executive summary that captures the essential elements investors need for quick decision-making.

**EXECUTIVE SUMMARY STRUCTURE:**

**1. Company Overview (2-3 sentences)**
- What the company does and its core value proposition
- Target market and customer segment
- Business model in simple terms

**2. Problem & Solution (3-4 bullet points)**
- Clear problem statement with market validation
- Unique solution and key differentiators
- Why now? Market timing and catalysts

**3. Market Opportunity**
- Market size and growth potential
- Target customer demographics
- Competitive landscape overview

**4. Business Model & Traction**
- Revenue model and pricing strategy
- Current traction metrics (users, revenue, partnerships)
- Key milestones achieved

**5. Team & Execution**
- Founder backgrounds and relevant experience
- Key team members and advisors
- Execution track record

**6. Financial Highlights**
- Current financial status
- Funding requirements and use of funds
- Revenue projections and path to profitability

**7. Investment Thesis**
- Key reasons to invest
- Potential risks and mitigation
- Exit potential and timeline

**OUTPUT FORMAT:**
Provide a well-structured 500-word executive summary using the framework above. Use bullet points, clear headings, and highlight key metrics and achievements. Focus on facts, numbers, and concrete evidence rather than marketing language.

**PITCH DECK TO ANALYZE:**
{pitch_text}
"""
        response = self.client.chat.completions.create(
            model="sonar-pro",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.5
        )
        summary = response.choices[0].message.content.strip()
        return {"summary": summary}

print("‚úÖ SummarizationAgent loaded exactly from your workspace")
print("üìù Creates executive summaries using Perplexity AI sonar-pro model")
````

````python
# FinancialAnalysisAgent: EXACT code from agents/financial_analysis_agent.py
class FinancialAnalysisAgent(BaseAgent):
    def __init__(self, perplexity_api_key: str):
        super().__init__("FinancialAnalysisAgent")
        self.client = OpenAI(
            api_key=perplexity_api_key,
            base_url="https://api.perplexity.ai"
        )

    def process(self, pitch_text):
        prompt = f"""
You are a senior financial analyst and venture capital partner with expertise in startup valuations and financial modeling. Conduct a thorough financial analysis of this startup based on the pitch deck.

**FINANCIAL ANALYSIS FRAMEWORK:**

1. **Revenue Analysis:**
   - Current revenue (if any) and historical growth
   - Revenue model analysis (subscription, transaction, advertising, etc.)
   - Revenue projections for next 3-5 years with assumptions
   - Unit economics: Customer Acquisition Cost (CAC), Lifetime Value (LTV)
   - Revenue diversification and concentration risks
   - Pricing strategy assessment

2. **Cost Structure & Burn Rate:**
   - Current monthly/annual burn rate
   - Cost breakdown (personnel, technology, marketing, operations)
   - Cost scalability and variable vs fixed costs
   - Path to profitability timeline
   - Capital efficiency metrics

3. **Financial Health & Runway:**
   - Current cash position and funding history
   - Estimated runway at current burn rate
   - Break-even analysis
   - Working capital requirements
   - Debt obligations and financial commitments

4. **Funding & Valuation:**
   - Previous funding rounds and valuations
   - Current funding requirements and use of funds
   - Comparable company valuations
   - DCF-based valuation estimate
   - Multiple-based valuation (Revenue, EBITDA multiples)
   - Dilution impact on existing shareholders

5. **Financial Risks:**
   - Revenue concentration risks
   - Cash flow volatility
   - Market dependency risks
   - Scalability financial challenges
   - Regulatory or compliance costs

6. **Investment Metrics:**
   - Return on investment projections
   - Payback period estimates
   - Risk-adjusted returns
   - Exit scenario analysis

**OUTPUT FORMAT:**
Provide a comprehensive 1500-word financial analysis covering all sections. Include calculations, assumptions, and benchmarking where possible. End with JSON summary:

{{
  "current_revenue": "...",
  "revenue_projection_y3": "...", 
  "monthly_burn_rate": "...", 
  "runway_months": "...", 
  "valuation_estimate": "...",
  "ltv_cac_ratio": "...",
  "break_even_timeline": "...",
  "funding_required": "...",
  "financial_health_score": "1-10",
  "key_financial_risks": [...]
}}

**PITCH DECK:**
{pitch_text}
"""
        # Fixed: single curly braces for dictionary
        response = self.client.chat.completions.create(
            model="sonar-pro",
            messages=[{"role": "user", "content": prompt}],
            temperature=0
        )
        financial_text = response.choices[0].message.content.strip()

        # Return a proper Python dictionary
        return {"financial_analysis": financial_text}

print("‚úÖ FinancialAnalysisAgent loaded exactly from your workspace")
print("üí∞ Analyzes revenue, burn rate, valuation using Perplexity sonar-pro")
````

````python
# MarketAnalysisAgent: EXACT code from agents/market_analysis_agent.py
try:
    from langchain.vectorstores import FAISS
    from langchain.embeddings import SentenceTransformerEmbeddings
    LANGCHAIN_AVAILABLE = True
except ImportError:
    LANGCHAIN_AVAILABLE = False
    print("‚ö†Ô∏è LangChain not available - install with: pip install langchain sentence-transformers faiss-cpu")

class MarketAnalysisAgent(BaseAgent):
    def __init__(self, perplexity_api_key: str):
        super().__init__("MarketAnalysisAgent")
        self.client = OpenAI(
            api_key=perplexity_api_key,
            base_url="https://api.perplexity.ai"
        )
        if LANGCHAIN_AVAILABLE:
            self.embedding_model = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
            self.vector_index = None

    def process(self, pitch_text):
        prompt = f"""
You are a senior market research analyst with 15+ years of experience analyzing startup markets and investment opportunities. Conduct a comprehensive market analysis based on the pitch deck provided.

**ANALYSIS FRAMEWORK:**

1. **Market Size & Opportunity:**
   - Total Addressable Market (TAM) - provide specific numbers and sources
   - Serviceable Addressable Market (SAM) 
   - Serviceable Obtainable Market (SOM)
   - Market growth trajectory over next 5 years
   - Geographic considerations and expansion potential

2. **Competitive Landscape:**
   - Direct competitors (at least 3-5 key players)
   - Indirect competitors and substitute products
   - Competitive positioning and differentiation factors
   - Market share distribution
   - Barrier to entry analysis

3. **Market Dynamics:**
   - Current growth rate with supporting data
   - Key market drivers and catalysts
   - Seasonal or cyclical patterns
   - Regulatory environment impact
   - Technology adoption trends

4. **Market Trends & Opportunities:**
   - Emerging trends affecting the market
   - Consumer behavior shifts
   - Technology disruptions
   - Regulatory changes
   - Partnership opportunities

5. **Challenges & Risks:**
   - Market saturation levels
   - Customer acquisition challenges
   - Competitive threats
   - Market timing risks
   - Economic sensitivity

**OUTPUT FORMAT:**
Provide a detailed 2000-word analysis covering all sections above. Include specific data points, statistics, and actionable insights. End with a JSON summary:

{{ 
  "market_size_tam": "...", 
  "market_size_sam": "...", 
  "growth_rate": "...", 
  "key_competitors": [...], 
  "market_maturity": "...", 
  "competitive_intensity": "...",
  "market_trends": [...],
  "key_challenges": [...],
  "opportunity_score": "1-10"
}}

**PITCH DECK:**
{pitch_text}
"""
        # Fixed: single curly braces for dictionary
        response = self.client.chat.completions.create(
            model="sonar-pro",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3
        )
        result_text = response.choices[0].message.content.strip()

        # Create vector index for semantic search
        if LANGCHAIN_AVAILABLE:
            self.vector_index = FAISS.from_texts([result_text], self.embedding_model)

        # Fixed: single curly braces for returning dict
        return {"market_analysis": result_text}

print("‚úÖ MarketAnalysisAgent loaded exactly from your workspace")
print("üìä Uses FAISS vector storage and Sentence Transformers for embeddings")
````

````python
# VCQuestionAnswerAgent: EXACT code from agents/vc_qa_agent.py
class VCQuestionAnswerAgent(BaseAgent):
    def __init__(self, openai_api_key: str, router=None):
        super().__init__("VCQuestionAnswerAgent")
        self.client = OpenAI(api_key=openai_api_key)
        self.router = router

    def generate_question(self, pitch_text: str) -> str:
        prompt = f"""
You are a VC analyst. Based on the pitch deck below, generate a relevant question a VC would ask.
Pitch Deck:
{pitch_text}

Return only the concise question.
"""
        response = self.client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.7
        )
        return response.choices[0].message.content.strip()

    def process(self, pitch_text: str, num_questions: int = 5) -> dict:
        qa_results = []

        for _ in range(num_questions):
            question = self.generate_question(pitch_text)

            if self.router:
                routing_text = f"Answer this question: {question}\nPitch: {pitch_text}"
                agent_name = self.router.decide_agent(routing_text)
                context = self.router.agents[agent_name].process(pitch_text)
            else:
                context = {"text": pitch_text}

            prompt = f"""
You are a VC analyst. Answer the question using the context.
Question: {question}
Context: {context}

Return JSON: {{ "question": "{question}", "answer": "..." }}
"""
            response = self.client.chat.completions.create(
                model="gpt-3.5-turbo",
                messages=[{"role": "user", "content": prompt}],
                temperature=0.3
            )
            answer_text = response.choices[0].message.content.strip()
            qa_results.append(answer_text)

        # Return proper Python dictionary
        return {"qa_history": qa_results}

print("‚úÖ VCQuestionAnswerAgent loaded exactly from your workspace")
print("‚ùì Generates autonomous questions and routes them via MCP for answers")
````

````python
# VerdictAgent: EXACT code from agents/verdict_agent.py
import json

class VerdictAgent(BaseAgent):
    def __init__(self, openai_api_key: str):
        super().__init__("VerdictAgent")
        self.client = OpenAI(api_key=openai_api_key)

    def process(self, agent_outputs: dict, vc_qa: dict) -> dict:
        """
        agent_outputs: dict containing outputs from all analysis agents
        vc_qa: dict containing VC Q&A history
        Returns: dict with verdict, confidence, investment amount, equity %, justification
        """
        prompt = f"""
You are an investment analyst AI. 
Based on the following startup analysis, provide:
1. Verdict: Invest / Maybe / Do Not Invest
2. Confidence score (0-1)
3. Recommended investment amount in USD
4. Recommended equity percent
5. Justification

Agent Outputs:
{json.dumps(agent_outputs, indent=4)}

VC Q&A History:
{json.dumps(vc_qa, indent=4)}

Return JSON only with keys: verdict, confidence, recommended_investment_usd, recommended_equity_percent, justification
"""
        response = self.client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3
        )

        answer_text = response.choices[0].message.content.strip()
        try:
            verdict_json = json.loads(answer_text)
        except json.JSONDecodeError:
            # fallback if GPT returns slightly invalid JSON
            verdict_json = {
                "verdict": "Unknown",
                "confidence": 0.0,
                "recommended_investment_usd": 0,
                "recommended_equity_percent": 0,
                "justification": answer_text
            }

        return verdict_json

print("‚úÖ VerdictAgent loaded exactly from your workspace")
print("‚öñÔ∏è Synthesizes all analysis + Q&A to make final investment decision")
````

````python
# MCPRouterAgent: EXACT code from agents/mcp_router_agent.py
class MCPRouterAgent(BaseAgent):
    def __init__(self, perplexity_api_key: str):
        super().__init__("MCPRouterAgent")
        self.client = OpenAI(
            api_key=perplexity_api_key,
            base_url="https://api.perplexity.ai"
        )

        # Initialize all analysis agents
        self.agents = {
            "summarization": SummarizationAgent(perplexity_api_key),
            "market": MarketAnalysisAgent(perplexity_api_key),
            "risk": RiskComplianceAgent(perplexity_api_key),
            "technical": TechnicalAnalysisAgent(perplexity_api_key),
            "financial": FinancialAnalysisAgent(perplexity_api_key),
        }

        if LANGCHAIN_AVAILABLE:
            self.embedding_model = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
            self.vector_index = None
            self.documents = []

    def process(self, task, data):
        if task in self.agents:
            output = self.agents[task].process(data)
            
            # Store in FAISS vector database
            if LANGCHAIN_AVAILABLE:
                self.documents.append(str(output))
                self.vector_index = FAISS.from_texts(self.documents, self.embedding_model)
            
            return output
        else:
            raise ValueError(f"Unknown task: {task}")

    def retrieve_context(self, query, top_k=3):
        """Retrieve relevant context from FAISS vector database"""
        if not LANGCHAIN_AVAILABLE or self.vector_index is None:
            return ""
        results = self.vector_index.similarity_search(query, k=top_k)
        return " ".join([r.page_content for r in results])

    def decide_agent(self, text):
        """Intelligent routing to determine which agent should handle the content"""
        prompt = f"""
You are an intelligent routing agent for a startup investment analysis system. Analyze the text and determine which specialized agent should handle this content based on the keywords, context, and focus area.

**AGENT SPECIALIZATIONS:**

- **summarization**: General overview, executive summary, company description, overall pitch content
- **market**: Market size, competition, industry analysis, customer segments, market trends, TAM/SAM/SOM
- **risk**: Legal issues, compliance, regulatory risks, operational risks, security concerns, liability
- **technical**: Technology stack, architecture, scalability, AI/ML, engineering, product development
- **financial**: Revenue, costs, burn rate, funding, valuation, financial projections, unit economics

**ROUTING RULES:**
1. Look for specific keywords and domain terminology
2. Consider the primary focus and intent of the content
3. If content spans multiple areas, choose the most prominent theme
4. Default to 'summarization' only for general company overviews

**ANALYSIS:**
Text content: {text}

Based on the content analysis, which agent should handle this?
Return ONLY the agent name in lowercase (summarization, market, risk, technical, financial).
"""
        response = self.client.chat.completions.create(
            model="sonar",
            messages=[{"role": "user", "content": prompt}],
            temperature=0
        )
        decision = response.choices[0].message.content.strip().lower()
        return decision if decision in self.agents else "summarization"

print("‚úÖ MCPRouterAgent loaded exactly from your workspace")
print("üîÄ Central orchestrator that routes requests and manages FAISS storage")
````

## üß† System Summary & Agent Explanations

### What Each Agent Does with MCP Integration:

**üîÄ MCP Router Agent** (Central Brain)
- **Role**: Acts as the central nervous system coordinating all agent interactions
- **MCP Function**: Routes pitch deck content to appropriate specialized agents
- **Intelligence**: Uses Perplexity AI to analyze content and decide optimal routing
- **Storage**: Stores all agent outputs in FAISS vector database for retrieval
- **Why Important**: Enables seamless multi-agent coordination and context sharing

**üìù Summarization Agent** 
- **Processing**: Creates executive summaries using Perplexity sonar-pro model
- **MCP Integration**: Receives raw pitch text from router, returns structured summary
- **Output**: Company overview, problem/solution, market opportunity, team highlights
- **Storage**: Summary vectors stored in FAISS for Q&A retrieval

**üí∞ Financial Analysis Agent**
- **Processing**: Deep financial modeling using Perplexity sonar-pro with calculation prompts
- **MCP Integration**: Analyzes financial sections routed by MCP Router
- **Output**: Revenue projections, burn rate, valuation estimates, investment metrics
- **Specialization**: Uses few-shot prompting for financial calculations and ratios

**üìä Market Analysis Agent** 
- **Processing**: Market research using Perplexity + FAISS vector storage
- **MCP Integration**: Handles all market-related queries routed by MCP
- **Data Sources**: Web scraping capabilities via Perplexity's real-time data
- **Storage**: Creates market analysis vectors with Sentence Transformers embeddings

**‚ö†Ô∏è Risk & Compliance Agent**
- **Processing**: Comprehensive risk assessment across legal, operational, financial dimensions
- **MCP Integration**: Processes risk-related content identified by router
- **Output**: Risk scores (1-10), mitigation strategies, compliance requirements
- **Expertise**: 20+ risk categories with probability and impact analysis

**üîß Technical Analysis Agent**
- **Processing**: Technology stack evaluation using Perplexity sonar-pro
- **MCP Integration**: Handles technical feasibility questions from router
- **Output**: Scalability assessment, innovation scores, technical risks
- **Focus**: Architecture, security, development practices, technical moat analysis

**‚ùì VC Question-Answer Agent**
- **Processing**: Generates 10 autonomous questions using OpenAI GPT-3.5-turbo
- **MCP Integration**: Routes each question to appropriate agent via MCP Router
- **Intelligence**: Uses retriever pattern to get context from FAISS storage
- **Workflow**: Question generation ‚Üí MCP routing ‚Üí Context retrieval ‚Üí Answer synthesis

**‚öñÔ∏è Investment Verdict Agent**
- **Processing**: Final investment decision using OpenAI GPT-3.5-turbo
- **Input**: All agent analyses + complete Q&A history
- **Output**: Invest/Maybe/Pass decision with confidence scores and reasoning
- **Integration**: Synthesizes all MCP-coordinated analysis into actionable verdict

1. **Coordinated Intelligence**: Each agent specializes while sharing context
